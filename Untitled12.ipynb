```python
# Smartphone Price Prediction & Feature Analysis

# This project analyzes smartphone specifications to:
# - Clean and preprocess raw data
# - Explore relationships between features
# - Use SQL for structured queries
# - Build machine learning models to predict price
# - Identify the most important features influencing price
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("uncleaned.csv")
df.head()
```

```python
print(df.columns.tolist())
df.info()
df.describe()
```

```python
# Drop empty rows
df.dropna(how='all', inplace=True)

# Convert price to numeric
df['price'] = pd.to_numeric(
    df['price'].astype(str).str.replace(r'[₹,]', '', regex=True),
    errors='coerce'
)

# Convert spec score
df['spec_score'] = pd.to_numeric(df['spec_score'], errors='coerce')

# Drop rows with missing critical values
df.dropna(subset=['price', 'spec_score'], inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
# Extract battery capacity (example: "5000 mAh")
df['battery_capacity'] = df['battery'].str.extract(r'(\d+)').astype(float)

# Extract RAM (example: "8GB")
df['ram'] = df['storage'].str.extract(r'(\d+)').astype(float)

# Extract display size
df['display_size'] = df['display_info'].str.extract(r'(\d+\.?\d*)').astype(float)

# Extract camera megapixels
df['camera_mp'] = df['camera'].str.extract(r'(\d+)').astype(float)

df.head()
```

```python
conn = sqlite3.connect("smartphones.db")
df.to_sql("phones", conn, if_exists='replace', index=False)
```

```python
# Most expensive phones
query1 = """
SELECT name, price
FROM phones
ORDER BY price DESC
LIMIT 10;
"""
pd.read_sql(query1, conn)
```

```python
# Average price by OS
query2 = """
SELECT os, AVG(price) as avg_price
FROM phones
GROUP BY os
ORDER BY avg_price DESC;
"""
pd.read_sql(query2, conn)
```

```python
plt.hist(df['price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['spec_score'], df['price'])
plt.xlabel("Spec Score")
plt.ylabel("Price")
plt.title("Spec Score vs Price")
plt.show()
```

```python
plt.scatter(df['battery_capacity'], df['price'])
plt.xlabel("Battery Capacity")
plt.ylabel("Price")
plt.title("Battery vs Price")
plt.show()
```

```python
features = ['spec_score', 'battery_capacity', 'ram', 'display_size', 'camera_mp']

df_model = df[features + ['price']].dropna()

X = df_model[features]
y = df_model['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

lr_preds = lr.predict(X_test_scaled)

print("LR MSE:", mean_squared_error(y_test, lr_preds))
print("LR R2:", r2_score(y_test, lr_preds))
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)

rf_preds = rf.predict(X_test)

print("RF MSE:", mean_squared_error(y_test, rf_preds))
print("RF R2:", r2_score(y_test, rf_preds))
```

```python
print("LR CV:", cross_val_score(lr, X_train_scaled, y_train, cv=10).mean())
print("RF CV:", cross_val_score(rf, X_train, y_train, cv=10).mean())
```

```python
importance = pd.Series(rf.feature_importances_, index=features)

importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```

```python
residuals = y_test - rf_preds

plt.scatter(rf_preds, residuals)
plt.axhline(y=0)
plt.title("Residual Plot")
plt.show()
```

```python
## Key Insights

#- Spec score strongly correlates with price
#- Battery and RAM also influence price
#- Random Forest performs better than Linear Regression
```
